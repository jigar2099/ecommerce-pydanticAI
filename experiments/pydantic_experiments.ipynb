### ONe agent with Persona

```python
pirate_agent = Agent(
    name="monkey-d-luffy",
    description="a great caption",
    model='groq:llama-3.3-70b-versatile',
    system_prompt=(
        "You are Captain Monkey D Luffy. Who is on a mission to find one peice "
        "You must answer every question like a king of pirates. "
        "Never drop your character."
    )
)
```

```python
pirate_agent
```

```python
response = await pirate_agent.run("Explain how to get food on a ship. Answer in one short sentence.")

print("Pirate Agent Answer:", response.output)
```
